```python
df = pd.read_excel('Cum. Intrusion vs. Pore Size.xlsx', skiprows=1)
```

```python
df.head()
```

```python
df.drop(df.columns[2], axis=1, inplace=True)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,6))
sns.jointplot(df, x=df['Pore size Diameter (µm)'], y=df['Cumulative Intrusion (mL/g)'])
```

```python
plt.figure(figsize=(10,6))
sns.jointplot(df, x=df['Pore size Diameter (µm)'], y=df['Log Differential Intrusion (mL/g)'])
```

```python
plt.figure(figsize=(10,6))
sns.pairplot(df)
plt.savefig('pairplot_intrusion vs. pore size.png')
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(df['Pore size Diameter (µm)'], df['Cumulative Intrusion (mL/g)'], color='orange', label='Cumulative Intrusion')
ax.set_xlabel('Pore size Diameter (µm)')
ax.set_ylabel('Cumulative Intrusion')
ax.legend()
plt.savefig('pore_size vs. cum. intrusion.png')
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(df['Pore size Diameter (µm)'], df['Log Differential Intrusion (mL/g)'], color='orange', label='Log Differential Intrusion')
ax.set_xlabel('Pore size Diameter (µm)')
ax.set_ylabel('Log Differential Intrusion')
ax.set_xscale('log')
ax.legend()
plt.savefig('pore_size vs. log. diff. intrusion.png')
```
